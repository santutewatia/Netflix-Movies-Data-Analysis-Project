categorizing Vote_Average column

We would cut the vote_Average values and make 4 categories: popular, average,  below_avg , and not_popular  to describe it more using catigorize_col() function  provided above



```python
def catigorizes(df,col,lables):
    edges=[df[col].describe()['min'],
          df[col].describe()['25%'],
          df[col].describe()['50%'],
          df[col].describe()['75%'],
          df[col].describe()['max']]
    df[col]=pd.cut(df[col],edges,labels=lables,duplicates='drop')
    return df 
```

```python
lables=['not_popular','below_avg','average','popular']

catigorizes(df,'Vote_Average',lables)

df['Vote_Average'].unique()
```
